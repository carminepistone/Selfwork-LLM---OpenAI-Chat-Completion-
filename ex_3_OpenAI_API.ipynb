**Traccia:** Crea un prompt che riassuma un articolo di notizie in un singolo paragrafo di al massimo 255 caratteri.

**Obiettivo:** Prendere un testo lungo e condensarlo nei punti essenziali.

```python
import os, openai
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv('.env')
openai.api_key = os.environ['OPENAI_API_KEY']
client = OpenAI()
```

```python
input_text = input("Inserisci l'articolo da riassumere: ")

completion = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "developer", "content": "Riassumi il testo in un singolo paragrafo di massimo 255 caratteri. Rispondi solo con il riassunto."},
        {"role": "user", "content": input_text}
    ],
)
print(completion.choices[0].message.content)
```
