**Obiettivo:** Sviluppare un prompt che sia adatto alla traduzione fluida tra italiano e inglese.

```python

# TODO Inizializza il client di Open AI
import os, openai
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv('.env') # Read the .env file

# Configure OpenAI client
openai.api_key = os.environ['OPENAI_API_KEY']

client = OpenAI()
```

```python
input_text = input("Inserisci il testo da tradurre: ")

completion = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "developer", "content": "Sei un traduttore. Se il testo è in italiano traducilo in inglese, se è in inglese traducilo in italiano. Rispondi solo con la traduzione."},
        {"role": "user", "content": input_text}
    ],
)
print(completion.choices[0].message.content)
```
