**Traccia:** Scrivi un prompt che generi domande di comprensione su un breve testo.

**Obiettivo:** Sviluppare un prompt per creare domande educative basate su un contenuto.

```python
import os, openai
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv('.env')
openai.api_key = os.environ['OPENAI_API_KEY']
client = OpenAI()
```

```python
input_text = input("Inserisci il testo: ")

completion = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "developer", "content": "Genera 5 domande di comprensione basate sul testo fornito. Rispondi con un elenco numerato di domande."},
        {"role": "user", "content": input_text}
    ],
)
print(completion.choices[0].message.content)
```
